```python
import numpy as np
import matplotlib as math
from matplotlib import pyplot as plt

# load histograms from files created with bash script
z, h1, h2, h3,h4,h5= np.loadtxt('hist1_5.csv',unpack=True)
h6, h7, h8, h9, h10 = np.loadtxt('hist6_10.csv',unpack=True)
h11, h12, h13, h14, h15 = np.loadtxt('hist11_15.csv',unpack=True)
h16, h17, h18, h19, h20, h21 = np.loadtxt('hist16_21.csv', unpack=True)

h = [h1,h2,h3,h4,h5,h6,h7,h8,h9,h10,h11,h12,h13,h14,h15,h16,h17,h18,h19,h20,h21]


def intersection_overlap(hist_1, hist_2): #computes overlap between histograms
    minima = np.minimum(hist_1, hist_2)
    intersection = np.true_divide(np.sum(minima), np.sum(hist_2))
    return intersection

def overlap_histogram(): #computes histograms of overlap
    dummy = []
    for i in range(len(h) - 3):
        a = np.minimum( (h[i] / np.sum(h[i])) , (h[i+3] / np.sum(h[i+3])) )
        dummy.append(a)
    return dummy

fig = plt.figure()
ax1 = fig.add_subplot(111)
ax1.set_xlabel(r'Distance along z $(nm)$')
ax1.set_ylabel(r'Frequency')

dummy = np.array(overlap_histogram())
for i in dummy:
    ax1.plot(z,i)
ax1.plot([],[],label = r'$3^{rd}$ neighbour'' overlap \nhistograms for hairpin',linewidth = 0)
leg = ax1.legend(loc = 'best')
plt.ylim([0,0.09])
plt.savefig('hairpin_overlap_third_NN.pdf')
plt.show()


'''
# define functions to calculate intersection & total area of two histograms
#testing to see whether it works for 2 histograms
mu_1 = 1 #mean of the first distribution
mu_2 = 4 #mean of the second distribution
sigma = 2.0
data_1 = np.random.normal(mu_1, 2.0, 10000)
data_2 = np.random.normal(mu_2, 2.0, 10000)
hist_1, _ = np.histogram(data_1, bins=100, range=[-15, 15])
hist_2, _ = np.histogram(data_2, bins=100, range=[-15, 15])

count1, bins1, ignored1 = plt.hist(data_1, 100, density=True)
count2, bins2, ignored2 = plt.hist(data_2, 100, density=True)
print (intersection_overlap(hist_1,hist_2))
plt.clf()
'''
#Percentage of "good" overlap

overlap = []
for i in range(len(h) - 3):
    overlap.append(intersection_overlap(h[i],h[i+3]))

bad =0
good = 0
for i in overlap:
    if i < 0.32:
        bad += 1
    else:
        good +=1
    total = float(good) / float(bad + good)
print ('Percentage: ', total)
print('Average: ', np.mean(overlap))
print('Standard Deviation: ',np.std(overlap))
```
